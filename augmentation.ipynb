```python
def rotate(image, angle, center = None, scale = 1.0):
    (h, w) = image.shape[:2]

    if center is None:
        center = (w / 2, h / 2)

    # Perform the rotation
    M = cv2.getRotationMatrix2D(center, angle, scale)
    rotated = cv2.warpAffine(image, M, (w, h))

    return rotated
```

```python
name="plate (182).bmp"
path="/content/drive/MyDrive/113db_project_8-8-16/"
save_path="/content/drive/MyDrive/test/"
img=cv2.imread(path+name)

plt.imshow(img, cmap=plt.cm.binary)
plt.grid(False)
plt.show()

img1 = cv2.flip(img, flipCode=-1)
plt.imshow(img1, cmap=plt.cm.binary)
plt.grid(False)
plt.show()

img2=rotate(img, 5)
plt.imshow(img2, cmap=plt.cm.binary)
plt.grid(False)
plt.show()

img3=rotate(img, 0, scale=0.9)
plt.imshow(img3, cmap=plt.cm.binary)
plt.grid(False)
plt.show()
```

```python
from skimage.util import random_noise

noise = random_noise(img, mode='s&p', amount=0.011)

# The above function returns a floating-point image in the range [0, 1]
# so need to change it to 'uint8' with range [0,255]
im = np.array(255 * noise, dtype=np.uint8)

plt.imshow(im, cmap=plt.cm.binary)
plt.grid(False)
plt.show()
```
